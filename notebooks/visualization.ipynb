## Univariate Analysis:

**Univariate Analysis:** Univariate analysis involves the study and examination of a single variable in a dataset to understand its distribution, central tendency, and variability.

### Exploratory Data Analysis: Continuous Features

```python
# Set the color palette with different shades of the theme color
custom_palette = ["#47525E", "#606B76", "#7A878F", "#949EA7", "#ADB7BF"]
```

## Bivariate Analysis:

**Bivariate Analysis:** Bivariate analysis involves the examination and exploration of the relationship between two variables in a dataset.

### Correlation Matrix:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(20, 20))
sns.heatmap(df.corr(), annot=True, cmap=custom_palette)  # Use custom_palette as the colormap
plt.show()
plt.show()
```

### Scatter Plots:


```python
custom_palette = ["#5C5470", "#94618E"]  # Using the theme color and a complementary color
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='EmpLastSalaryHikePercent', hue='Attrition', palette=custom_palette)
plt.title("Scatter Plot of Age vs EmpLastSalaryHikePercent", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("EmpLastSalaryHikePercent", fontsize=12)
plt.legend(title="Attrition")
plt.show()
```

## Multivariate Analysis:

**Multivariate Analysis:** Multivariate analysis encompasses the study of relationships and patterns among multiple variables within a dataset, enabling a comprehensive understanding of complex interactions and correlations.

### Pairplot:

```python
sns.pairplot(df, vars=['Age', 'EmpLastSalaryHikePercent', 'ExperienceYearsInCurrentRole'], hue='Attrition', palette=custom_palette)
plt.suptitle("Pairplot of Selected Continuous Variables", y=1.02, fontsize=16)
plt.show()
```
